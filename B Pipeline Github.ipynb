```python
# As usual importing the necessary libraries is the first zero. For what I want to achieve, I only need the pandas library.
import pandas as pd 
```

```python
# as usual importing the data in is step one. The file is under my downloads and the specific worksheet I want info. from is named Speedings
client_data=pd.read_excel(r'C:\Users\glen.ochieng\Downloads\Client B May Overspeeding Data.xlsx',sheet_name='Speedings')
```

```python
# Checking if the dataset has all the columns I want 
client_data.columns
```

```python
# Removing the 'Grouping' column as its a duplicate of the 'Reg' column using the drop method
# Also note that the Beginning columnn contains date time information of every trip a bike took along with the speed which is our main variable.
client_data = client_data.drop(columns=['Grouping'])
```

```python
client_data
```

```python
# now to check the data types these columns came in
client_data.dtypes
```

#### Here was my main pain! The 'Beginning' came in in the general (excel) datatype or object (python) for some values
#### and   date value for others. In short it came in as mixed in stead of everything as a date time data type. 

```python
# Luckily Python has a code that can fix a column that has mixed data type entries and simultaneously transforms everything 
# date-time data type that I want.

# All it takes is using the pandas method : pd.to_datetime and including the arguement format='mixed'
# Including errors='coerce' is a btw . It will bring NaN for values that can't be converted and I can spot them more easily in case they are any.

client_data['Beginning']=pd.to_datetime(client_data['Beginning'],format='mixed',errors='coerce')
```

```python
client_data.dtypes
```

#### Excellent! The 'Beginning' column is now in the proper format.

```python
client_data
```

```python
# I also wish to extract just the date from the 'Beginning' column  and store them into a column named 'Date'
# We can use the method dt.strftime which can extract dates from a date-type variable.
# Additionally, we can even select the format the date should be in. My raw data came in as year-month-date
# and I want it to be day-month- year for ease of understanding. 

# Con is it will convert not convert it back to a date/date type variable

client_data['Date'] = client_data['Beginning'].dt.strftime('%d-%m-%Y')
```

```python
client_data
```

```python
# just checking to ensure the date column in which data type
client_data.dtypes
```

```python
# Same as code as before to convert'Date' to the correct data type.
client_data['Date'] = pd.to_datetime(client_data['Date'], format='%d-%m-%Y')
```

```python
# I also wish to extract time from the 'Beginning' column using .dt.time
client_data['Time'] = client_data['Beginning'].dt.time
```

```python
client_data
```

```python
client_data.dtypes
```

```python
# My client is only intrested in speeds above 80 km/h as the speed limit for commercial vehicles/bikes in Kenya is 80 km/h. 
# As such, I apply a filter to only include bikes with speeds of above 80 km/h.

client_data=client_data[client_data['Max speed']>80]
```

```python
client_data
```

```python
# Now that I have made new columns, I can shed some of the old ones
client_data=client_data[['Reg','Location','Date','Time','Max speed']]
```

```python
# Saving the formatted dataset back in my Downloads folder and naming it Client B cleaned data
client_data.to_excel(r'C:\Users\glen.ochieng\Downloads\Client B Data.xlsx', index=False, engine='openpyxl')
```
